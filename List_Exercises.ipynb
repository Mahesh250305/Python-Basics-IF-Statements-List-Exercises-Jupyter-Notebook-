### 1. Remove Extreme Values from List

This program:

- Takes a list of numbers from the user
- Removes:
  - n smallest values
  - n largest values
- Returns the new list

Conditions:
- n = 2 (as per question)
- If list has less than 4 elements → show error

```python
def remove_extremes(lst, n):
    new_list = sorted(lst)
    return new_list[n: len(new_list) - n]

# Main program
user_input = input("Enter numbers separated by space: ")
numbers = list(map(int, user_input.split()))

if len(numbers) < 4:
    print("Error: Enter at least 4 numbers")
else:
    result = remove_extremes(numbers, 2)
    print("List after removing extremes:", result)
    print("Original list:", numbers)
```

### 2. Extract Words from String

This program:

- Takes a string input
- Removes punctuation from edges of words
- Keeps punctuation inside words (like don't, isn't)
- Returns list of clean words

```python
def clean_words(text):
    words = text.split()
    clean_list = []

    for word in words:
        # Remove punctuation from start and end only
        while len(word) > 0 and not word[0].isalnum():
            word = word[1:]
        while len(word) > 0 and not word[-1].isalnum():
            word = word[:-1]
        
        if word != "":
            clean_list.append(word)
    
    return clean_list


# Main program
text = input("Enter a sentence: ")
result = clean_words(text)

print("Clean words:", result)
```
