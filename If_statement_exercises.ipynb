### 1. Check Even or Odd

This program takes an integer input from the user and checks whether it is even or odd using an if statement.

```python
num = int(input("Enter an integer: "))

if num % 2 == 0:
    print("The number is Even")
else:
    print("The number is Odd")
```

### 2. Human Years to Dog Years

- First 2 years = 10.5 dog years each  
- After that = 4 dog years per year  
- Negative input shows error

```python
human_years = float(input("Enter human years: "))

if human_years < 0:
    print("Error: Age cannot be negative")
else:
    if human_years <= 2:
        dog_years = human_years * 10.5
    else:
        dog_years = (2 * 10.5) + (human_years - 2) * 4
    
    print("Dog years:", dog_years)
```

### 3. Check Vowel or Consonant

- a, e, i, o, u → vowel

### 4. Shape Name from Number of Sides

Enter number of sides (3–10) and display the shape name.

```python
sides = int(input("Enter number of sides: "))

if sides < 3 or sides > 10:
    print("Error: Enter sides between 3 and 10")
else:
    if sides == 3:
        print("Triangle")
    elif sides == 4:
        print("Quadrilateral")
    elif sides == 5:
        print("Pentagon")
    elif sides == 6:
        print("Hexagon")
    elif sides == 7:
        print("Heptagon")
    elif sides == 8:
        print("Octagon")
    elif sides == 9:
        print("Nonagon")
    elif sides == 10:
        print("Decagon")
```

### 6. Sound Level Classification

This program reads a sound level in decibels from the user and:

- Displays exact noise if it matches:
  - 130 → Jackhammer
  - 106 → Gas lawnmower
  - 70 → Alarm clock
  - 40 → Quiet room

- If the value lies between two levels, it shows the range.

- Also handles:
  - Values below 40
  - Values above 130

```python
db = int(input("Enter sound level in decibels: "))

if db == 130:
    print("Jackhammer")
elif db == 106:
    print("Gas lawnmower")
elif db == 70:
    print("Alarm clock")
elif db == 40:
    print("Quiet room")

elif db > 106 and db < 130:
    print("Between Gas lawnmower and Jackhammer")
elif db > 70 and db < 106:
    print("Between Alarm clock and Gas lawnmower")
elif db > 40 and db < 70:
    print("Between Quiet room and Alarm clock")

elif db < 40:
    print("Quieter than a Quiet room")
elif db > 130:
    print("Louder than a Jackhammer")

else:
    print("Invalid input")
```

### 7. Music Note Frequency

This program:

1. Takes a note as input (e.g., C4, A4, G5)
2. Uses base frequencies of 4th octave:
   - C4 = 261.63
   - D4 = 293.66
   - E4 = 329.63
   - F4 = 349.23
   - G4 = 392.00
   - A4 = 440.00
   - B4 = 493.88

3. Adjusts frequency using octave formula:
   - frequency = base × 2^(octave - 4)

4. Works for notes from C0 to C8

```python
note = input("Enter note (e.g., C4, A4): ").upper()

# Extract note and octave
letter = note[0]
octave = int(note[1])

# Base frequency for 4th octave
if letter == 'C':
    base = 261.63
elif letter == 'D':
    base = 293.66
elif letter == 'E':
    base = 329.63
elif letter == 'F':
    base = 349.23
elif letter == 'G':
    base = 392.00
elif letter == 'A':
    base = 440.00
elif letter == 'B':
    base = 493.88
else:
    print("Invalid note")
    base = None

# Calculate final frequency
if base is not None:
    freq = base * (2 ** (octave - 4))
    print("Frequency:", round(freq, 2), "Hz")
```

### 8. Frequency to Musical Note

This program:

- Takes a frequency (Hz) from the user
- Compares it with known frequencies (C4–B4)
- If within ±1 Hz → prints the note name
- Otherwise → prints "No matching note"

```python
freq = float(input("Enter frequency (Hz): "))

if abs(freq - 261.63) <= 1:
    print("C4")
elif abs(freq - 293.66) <= 1:
    print("D4")
elif abs(freq - 329.63) <= 1:
    print("E4")
elif abs(freq - 349.23) <= 1:
    print("F4")
elif abs(freq - 392.00) <= 1:
    print("G4")
elif abs(freq - 440.00) <= 1:
    print("A4")
elif abs(freq - 493.88) <= 1:
    print("B4")
else:
    print("Frequency does not correspond to a known note")
```

### 9. Wavelength to Color

This program:

- Takes wavelength in nanometers (nm)
- Determines color based on range:

380–<450 → Violet  
450–<495 → Blue  
495–<570 → Green  
570–<590 → Yellow  
590–<620 → Orange  
620–750 → Red  

- Shows error if outside visible spectrum

```python
wavelength = float(input("Enter wavelength (nm): "))

if wavelength >= 380 and wavelength < 450:
    print("Violet")
elif wavelength >= 450 and wavelength < 495:
    print("Blue")
elif wavelength >= 495 and wavelength < 570:
    print("Green")
elif wavelength >= 570 and wavelength < 590:
    print("Yellow")
elif wavelength >= 590 and wavelength < 620:
    print("Orange")
elif wavelength >= 620 and wavelength <= 750:
    print("Red")
else:
    print("Error: Wavelength is outside the visible spectrum")
```
